```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("titanic.csv")
df
```

```python
df.info()
```

```python
cabin_unique_count = df['Cabin'].nunique()
embarked_unique_count = df['Embarked'].nunique()
print(f"Cabin unique count: {cabin_unique_count} and Embarked unique count: {embarked_unique_count}")
```

Bar Chart


```python
embarked_unique_count = df['Embarked'].nunique()
print(f"Embarked unique count: {embarked_unique_count}")
embarked_counts = df['Embarked'].value_counts()

plt.bar(embarked_counts.index, embarked_counts.values)
plt.show()
```

```python
plt.bar(embarked_counts.index, embarked_counts.values, color=['blue', 'green', 'orange'])
plt.xlabel("Embarked")
plt.ylabel("Count")
plt.title("Count of Passengers by Embarked Port")
plt.show()
```

Pie Chart


```python
Sex_counts = df['Sex'].value_counts()

plt.pie(Sex_counts)
plt.show()
```

```python
explode = (0, 0.2)

fig, ax = plt.subplots()
ax.pie(Sex_counts, colors=['blue', 'red'], explode=explode, labels=Sex_counts.index, autopct='%1.1f%%',shadow=True, radius=1.5)
plt.show()
```

Histogram


```python
plt.hist(df.Age)
plt.show()
```

```python
plt.hist(df.Age, bins=50, edgecolor='black')
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.title("Distribution of Age Histogram")
plt.show()
```

ScatterPlot


```python
plt.scatter(df.Age, df.Fare)
plt.show()
```

```python
# Current value of each ticket in USD
CPI_1912 = 9.8
CPI_2025 = 319.082  # January 2025 CPI
EXCHANGE_RATE_2025 = 1.268  # USD/GBP forecast for May 2025

# Inflation adjustment factor
inflation_factor = CPI_2025 / CPI_1912

# Function to adjust fare to 2025 USD
def adjust_fare_to_2025_usd(fare_1912_gbp):
    # Adjust for inflation
    fare_2025_gbp = fare_1912_gbp * inflation_factor
    # Convert to USD
    fare_2025_usd = fare_2025_gbp * EXCHANGE_RATE_2025
    return fare_2025_usd

# Apply the function to the DataFrame
df['Fare_2025_USD'] = df['Fare'].apply(adjust_fare_to_2025_usd)

# Display the DataFrame
df
```

```python
plt.figure(figsize=(8, 5))
plt.scatter(df['Age'], df['Fare_2025_USD'], color='blue', alpha=0.6, edgecolors='black')
plt.xlabel("Age of Passenger", fontsize=12)
plt.ylabel("Ticket Fare in 2025 USD ($)", fontsize=12)
plt.title("Titanic Ticket Prices Adjusted for 2025 Inflation", fontsize=14)
plt.show()
```
