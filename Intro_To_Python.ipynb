A **while loop** we is used to execute a set of statements as long as a condition is true

```python
i=1
while i<9:
  print(i)  #beware of the indentation
  i += 1

print('we are out of the while loop')
```

```python
# Calculate the sum of numbers from 1 to 10
total = 0
num = 1
while num <= 10:
    total += num # gives us total = 1 (on the first loop)
    num += 1 #num= num +1
print('The sum of the numbers from 1 to 10 is', total)
```

```python
# How do I print out average?
average= total/10
print('average is', average)
```

# Control conditions

**if** statements executes a block of code only if a specified condition is true.

```python
x = 300
y = 200

if(x >= y):
  print('x is greater than or equals to y') #beware of the indentation

if(x <= y):
  print('x is not greater than or equals to y')
```

**else** statements catches anything which isn't caught by the preceding conditions

```python
y=8
if (y==7):
  print('y is seven')
else:
  print('y is not 7, y is', y)
```

**elif** statments are used when a  previous conditions is not true to then try another condition

```python
x=10

if (x==5):
  print('x is five')
elif (x==8):
  print('x is eight')
elif(x==10):
  print('x is ten')
else:
  print('x is not 5 nor 8 nor 10')
```

**and** is used to combine conditional statements where both conditons are true



```python
a = 200
b = 33
c = 500
if a > b and c > a:
  print("Both conditions are True")
```

**or** is used to combine conditional statements where at least one condition is true

```python
a = 200
b = 33
c = 500
if a > b or a > c:
  print("At least one of the conditions is True")
```

**not** is used to reverse ae conditional statement

```python
a = 33
b = 200
if not a > b:
  print("a is NOT greater than b")
```

```python
decision = input("Truth or Dare:")

if (decision=="Truth"):
  print('Tell a story')
elif (decision=="Dare"):
  print('I dare you to follow the MLDS instagram')
else:
  print('invalid')
```

# Advanced data types

**Lists** are used to store multiple items in a single variable

Lists are:

*   Ordered
*   Mutable(Elements can be changed)
*   Allow for Duplicate
*   Can store elements of different data types
*   Indexed (starting from 0 for the first element and -1 for the last)

```python
#A list with various data types
mixed_list = ['apples', 1 , 2+2 , 3.14, True]
print(mixed_list)
```

```python
# Creating a list
my_list = [1, 2, 3, 4, 5]
print(my_list)

#Indexing
print(my_list[0])
print(my_list[-1])
```

```python
#Indexing multiple elements
print(my_list[0:3])
```

```python
#Duplcating list

copy_list = my_list.copy()
print(copy_list)
```

```python
# Modifying elements
my_list[0] = 10
print(my_list)
```

**len** is used to get the length of a list

```python
len(my_list)
```

```python
# Appending elements
my_list.append(6)
print(my_list)

len(my_list)
```

```python
# Removing elements
my_list.remove(3)
print(my_list)

len(my_list)
```

```python
#Lets compare the orignal list to the changed list

print(copy_list)
print(my_list)
```

Notice how they are the same even though we did not change the copied list, that is because we created what is know as a shallow copy

A shallow copy makes a new object but only copies references to the inner objects, so changes to shared inner objects affect both copies

A deep copy makes a completely independent clone of the object and all its inner objects, so changes in one do not affect the other

**Tuples** are very similar to lists the main difference is that they are **immutable** meaning they can not be changed

```python
# Creating a tuple
my_tuple = (1, 2, 3, 4, 5)

len(my_tuple)
```

```python
x=my_tuple[0]

# Accessing elements
print(x)
```

```python
# Tuples are immutable - this would raise an error:
#my_tuple[0] = 10
```

**Sets** are also like lists however they are unordered, unchangeable, and can not be copied

```python
thisset = {"apple", "banana", "cherry", "apple"}

print(thisset)
```

Combine List with For Loop

```python
# Define a list
fruits = ["apple", "banana", "cherry", "orange"]
print(fruits)

# Iterate over each item in the list and print them
for fruit in fruits:
  print(fruit)
```

```python
# Iterating over a tuple
my_tuple = (1, 2, 3, 4, 5)
for item in my_tuple:
    print(item)
```

```python
# Iterating over a set
my_set = {1, 2, 3, 4, 5}
for item in my_set:
    print(item)
```
