#1. Making a basic Request

```python
import requests                     #sends HTTP requests to APIs
import json                         #handles JSON data format (text → Python objects)
import pandas as pd                 #turns JSON into tables (DataFrames)
import matplotlib.pyplot as plt     #lets us plot data later
```

```python
# This URL is an API endpoint and acts like a "data server" on the internet
url = "https://jsonplaceholder.typicode.com/posts/1"


response = requests.get(url) #By sending a HTTP GET request → Receive Data

data = response.json()

data
```

# 2. Working With JSON Data

JSON is derived from JavaScript (JavaScript Object Notation) and is just the format using key value pairs. Values can be booleans, strings, numbers, arrays or objects. That's all it is.

```python
#APIs usually return JSON which looks like a dictionary
data = response.json()
data
```

The API returned a Python dictionary that came from JSON.

```python
# Each key (like "name" or "age") is a column in the data.

name = data["name"]     # Get the value associated with the "name" field
age = data["age"]       # Get the predicted age from the API
count = data["count"]   # Get how many samples were used to make the prediction

print(f"Name: {name}")
print(f"Predicted Age: {age}")
print(f"Sample Size: {count}")
```

```python
#Changing the response with a different URL
name = "sophia"
url = f"https://api.agify.io?name={name}"

data = requests.get(url).json()
data
```

#3. Using Pokémon Data

```python
url = "https://pokeapi.co/api/v2/pokemon/pikachu"
response = requests.get(url)
data = response.json()

data["name"]
```

```python
height = data["height"]
weight = data["weight"]

print("Height:", height)
print("Weight:", weight)
```

Some values are inside lists and dictionaries.

```python
abilities = []

for ability in data["abilities"]:
    abilities.append(ability["ability"]["name"])

abilities
```

**APIs don't always work:**
* Internet issues

* Bad URLs

* Server problems

**Always check the status code.**

```python
url = "https://pokeapi.co/api/v2/pokemon/invalidpokemon"
response = requests.get(url)

if response.status_code == 200:
    data = response.json()
    print(data["name"])
else:
    print("Error:", response.status_code)
```

#4. Mini Project

**Write a Python script that:**

Calls an API

Extracts at least 3 values

Prints a clean summary

```python
#Pokémon data
pokemon = "charizard"
pokemon_url = f"https://pokeapi.co/api/v2/pokemon/{pokemon}"
```

```python
#Step 1 Select a Pokémon
pokemon = input("Enter a Pokémon name: ").lower()
```

```python
#Step 2 Build the API URL
url = f"https://pokeapi.co/api/v2/pokemon/{pokemon}"
```

```python
#Step 3 Send the API Request
response = requests.get(url)
```

```python
#Step 4 Handle Errors and Parse the JSON Response
if response.status_code != 200:
    print("❌ Pokémon not found. Please check the spelling.")
else:
    data = response.json()
```

```python
#Step 5 Extract Data (3+ Values)
name = data["name"].title()
height = data["height"]
weight = data["weight"]

abilities = []
for ability in data["abilities"]:
      abilities.append(ability["ability"]["name"])
```

```python
#Step 6 Display a Clean Summary
print("\n Pokémon Summary")
print("-------------------")
print(f"Name: {name}")
print(f"Height: {height}")
print(f"Weight: {weight}")
print("Abilities:", ", ".join(abilities))
```
